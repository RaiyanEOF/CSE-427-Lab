# Implementing Adaboost from Scratch
##Steps:
### ***1. Assign Weights to the Dataset***
- Start by giving **equal weight** to every sample:  
  $w_i = \frac{1}{N} \quad \text{for } i = 1, 2, ..., N$

### ***2. Create Stumps (Weak Learners)***
- Use **Decision Stumps**, i.e., Decision Trees with `max_depth = 1`.
- Use **Gini Index** to find the best split at each step.

### ***3. For Each Iteration (Round of Boosting)***

#### a. **Train the Weak Learner**
- Fit a stump on the training data using the current sample weights.

#### b. **Predict and Compute Error**
- Calculate the weighted classification error:
  $$
  \text{error}_t = \sum_{i=1}^{N} w_i \cdot \mathbb(h_t(x_i) \ne y_i)
  $$
  Here,
  
  $w_i$ is the weight of instance $i$
  
  $h_t(x_i)$ is the prediction of weak classifier at iteration $t$

  $y_i$ is the true label

  $(h_t(x_i) \ne y_i)$ returns 1 if the condition is true and 0 otherwise

#### c. **Compute the "Amount of Say" (Alpha)**
- This is the weight given to the weak learner:
  $$
  \alpha_t = \eta \cdot \frac{1}{2} \log\left(\frac{1 - \text{error}_t}{\text{error}_t}\right)
  $$
- Where:
  - $\eta$ is the **learning rate** (e.g., 0.1, 0.5, 1.0)

### ***4. Modify the Sample Weights***

- **Increase** the weights of **incorrectly classified** samples:
  $$
  w_i \leftarrow w_i \cdot e^{\alpha_t}
  $$

- **Decrease** the weights of **correctly classified** samples:
  $$
  w_i \leftarrow w_i \cdot e^{-\alpha_t}
  $$

- Combined in one expression:
  $$
  w_i \leftarrow w_i \cdot e^{\alpha_t \cdot (2 \cdot(y_i \ne h_t(x_i)) - 1)}
  $$

### ***5. Normalize the Weights***

Ensure the weights sum to 1:
$$
w_i \leftarrow \frac{w_i}{\sum_{j=1}^N w_j}
$$

### ***6. Compute the Cumulative Weights***

Convert the normalized weights into cumulative weights / probabilities:

$$
C_i = \sum_{j=1}^{i} w_j
$$

where:

- $C_i$ is the cumulative weight of sample $i$.
- The last cumulative weight will always be 1 (or very close to 1).

These cumulative weights define the probability intervals used for selecting samples for the next iteration.

### ***7. Generate Random Values***

Generate $N$ random numbers between 0 and 1, where $N$ is the number of training samples:

$$
r_i \sim U(0,1)
$$

These random values determine which samples will be selected for the next training dataset.

### ***8. Match Random Values with the Cumulative Weights***

For each random value, select the **first cumulative weight** that is greater than or equal to that random value.

This process performs **weighted random sampling**, meaning:

- Samples with **higher weights** (usually the incorrectly classified ones) have a higher probability of being selected multiple times.
- Samples with **lower weights** may be selected fewer times or not at all.

The selected samples form the new training dataset for the next boosting iteration.

### ***9. Final Prediction (Ensemble Output)***

The final prediction is a **weighted vote** of all weak learners:

$$
F(x) = \sum_{t=1}^{T} \alpha_t \cdot h_t(x)
$$

Classify based on the sign:

$$
\hat{y} = \text{sign}(F(x))
$$

### Notes:
- Ensure labels are in $\{-1, +1\}$ for proper calculations.

```python
from sklearn.tree import DecisionTreeClassifier # To create the stumps

class AdaBoostClassifier:
    def __init__(self, n_estimators, learning_rate):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.alphas = []
        self.models = []

    def fit(self, X, y):
        np.random.seed(42)

        X = np.array(X)
        y = np.array(y)

        n_samples, _ = X.shape

        # assign initial weights to the dataset
        weights = np.ones(n_samples) / n_samples

        X_current = X.copy()
        y_current = y.copy()

        # For each iteration
        for estimator in range(self.n_estimators):
            # 1. create a stump
            stump = DecisionTreeClassifier(
                max_depth=1,
                criterion='gini',
                random_state=42
            )

            # 2. train decision stump on the initial/resampled dataset
            stump.fit(X_current, y_current)

            # 3. predict on the training dataset
            predictions = stump.predict(X)

            # 4. compute error
            error = np.sum(weights * (predictions != y))

            # 5. check error and compute amount of say
            error = max(error, 1e-10)
            error = min(error, 0.999999)

            alpha = (self.learning_rate * 0.5 * np.log((1 - error) / error))

            # 6. update weights
            weights = (weights * np.exp(alpha * (2 * (predictions != y) - 1)))

            # 7. normalize weights
            weights = weights / np.sum(weights)

            # 8. compute cumulative weights
            cumulative_weights = np.cumsum(weights)

            # 9. generate random numbers
            random_values = np.random.uniform(0,1,n_samples)

            # 10. select rows where random value matches its cumulative weight range
            selected_indices = np.searchsorted(cumulative_weights,random_values)

            # 11. create new training dataset for next iteration
            X_current = X[selected_indices]
            y_current = y[selected_indices]

            # Save model and alpha
            self.models.append(stump)
            self.alphas.append(alpha)

    def predict(self, X):
        X = np.array(X)

        final_pred = np.zeros(X.shape[0])

        # Loop through all the stumps and calculate the final prediction
        for alpha, model in zip(self.alphas, self.models):
            final_pred += alpha * model.predict(X)

        final_pred = np.sign(final_pred)
        final_pred[final_pred == 0] = 1

        return final_pred.astype(int)



# Convert labels to {-1, 1} instead of {0,1}
y_train_ab = y_train.replace({0: -1, 1: 1})
y_test_ab = y_test.replace({0: -1, 1: 1})

# Train AdaBoost from scratch
adaboost = AdaBoostClassifier(n_estimators=50, learning_rate = 1)
adaboost.fit(X_train, y_train_ab.to_numpy())

# Predict and evaluate
y_pred_ab = adaboost.predict(X_test)

print("AdaBoost Classifier (from scratch) Accuracy:", accuracy_score(y_test_ab, y_pred_ab))
```
