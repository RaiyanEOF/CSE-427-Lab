## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Part A : Numpy

## 1. Create two arrays

```python
a = np.random.randint(1,101,20)
b = np.random.randint(1,101,20)
print("Array A:",a)
print("Array B:",b)
```

## 2. Convert into 5×4 matrices

```python
a = a.reshape(5,4)
b = b.reshape(5,4)
print("Matrix A:\n ",a)
print("Matrix B:\n ",b)
```

## 3. Scalar Multiplication

```python
a = a*10
b = b*25
print("Matrix A after scalar multiplication:\n",a)
print("Matrix B after scalar multiplication:\n",b)
```

## 4. Modify Matrix B and Perform Matrix Multiplication

```python
b = b.T
c = np.matmul(a,b)
print("Matrix C:\n",c)
```

## 5. Print Column 2 and 3 using Slicing

```python
print(c[:,1:3])
print(c[:])
```

## 6. Largest Element and Index

```python
largest = np.max(c)
print("Largest Value =",largest)
row,col = np.unravel_index(np.argmax(c), c.shape)
print("Index = " ,row,col)
```

## 7. Convert Matrix C into One-Dimensional Array

```python
c_1D = c.flatten()
print(c_1D)
```

## Part C: Matplotlib


## 1. Scatter Plot

```python
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28,25,29,20,26,50,19,29,52,55]

plt.figure(figsize=(6,5))
plt.scatter(temperature_dhaka,humidity_dhaka,marker="*",color="red")

plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Temperature vs Humidity")
plt.show()
```

## 2. Line Plot

```python
study_hours = [2,3,4,4,5,6,7,7,8,9,9,10,11,11,12]
marks = [6,10,15,20,34,44,55,60,55,67,70,80,90,99,100]

plt.figure(figsize=(12,8))
plt.plot(study_hours,marks)

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

## 3. Horizontal Bar Graph

```python
subjects = ['Maths','English','Science','Physics','Computer']
marks = [89,90,45,78,99]

plt.figure(figsize=(7,5))
plt.barh(subjects,marks)

plt.xlabel("Marks")
plt.title("Subject Scores")
plt.show()
```

## 4. Vertical Bar Graph

```python
colors = ['red','green','blue','orange','purple']

plt.figure(figsize=(7,5))
plt.bar(subjects,marks,color=colors)

plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Subject Scores")
plt.show()
```

## 5. All Graphs Using Subplots

```python
#Scatter Plot
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28,25,29,20,26,50,19,29,52,55]

#Line Plot
study_hours = [2,3,4,4,5,6,7,7,8,9,9,10,11,11,12]
marks = [6,10,15,20,34,44,55,60,55,67,70,80,90,99,100]

#Bar
subjects = ['Maths','English','Science','Physics','Computer']
subject_marks = [89,90,45,78,99]

plt.figure(figsize=(10,8))

#Scatter Plot
plt.subplot(2,2,1)
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.title("Scatter Plot")

#Line Plot
plt.subplot(2,2,2)
plt.plot(study_hours, marks)
plt.title("Line Plot")

#Horizontal Bar Plot
plt.subplot(2,2,3)
plt.barh(subjects, subject_marks)
plt.title("Horizontal Bar Graph")

#Vertical Bar Plot
plt.subplot(2,2,4)
plt.bar(subjects, subject_marks,color=['red','green','blue','orange','purple'])
plt.title("Vertical Bar Graph")
plt.tight_layout()
plt.show()
```
